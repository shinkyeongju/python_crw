### 정규 표현식으로 스크레이핑

- 문자 클래스로 만들어진 정규식은 '[] 사이의 문자들과 매치'라는 의미
- 문자 클래스([]) 안에는 어떤 문자나 메타 문자도 사용할 수 있음
- 문자 클래스 안에 ^ 메타 문자를 사용할 경우에는 반대(not)라는 의미

자주 사용하는 문자 클래스
- \d : 숫자와 매치, [0-9]와 동일한 표현식
- \d : 숫자가 아닌 것과 매치, [^0-9]와 동일한 표현식
- \s : whitespace 문자와 매치, [\t\n\r\f\v]와 동일한 표현식
- \S : whitespace 문자가 아닌 것과 매치, [^\t\n\r\f\v]와 동일한 표현식
- \w : 문자+숫자와 매치, [a-zA-Z0-9]와 동일한 표현식
- \W : 문자+숫자가 아닌 것과 매치, [^a-zA-Z0-9]와 동일한 표현식

Dot(.)
- 줄바꿈 문자인 \n을 제외한 모든 문자와 매치됨을 의미

반복 (*)
- (*) 바로 앞에 있는 문자가 0부터 무한대로 반복될 수 있다는 의미
- ex) ca*t --> 문자 a가 0부터 무한대로 반복 가능

반복 (+)
- (+) 바로 앞에 있는 문자가 최소 1번이상 반복될 수 있다는 의미

반복 ({m,n})
- {m,n} 앞에 있는 문자가 m부터 n까지 반복될 수 있다는 의미
- {3,} : 반복 횟수 3 이상, {,3} : 반복 횟수 3 이하

반복 (?)
- {0,1}을 의미

그루핑
- ()을 사용
- (ABC)+ : ABC 문자열이 계속해서 반복

http://www.hanbit.co.kr/store/books/full_book_list.html 에서 url 가져오기

```python
import re
from urllib.request import urlopen
from html import unescape

f = urlopen('http://www.hanbit.co.kr/store/books/full_book_list.html')
edcoding = f.info().get_content_charset(failobj='utf-8')
text = f.read().decode(encoding)

for partial_html in re.findall('<td class="left"><a.*</td>',text): # text안에서 '<td> ~ '를 찾으라는 뜻 
    url = re.search('<a href="(.*)">', partial_html).group(1)   # group(0)과 group(1)의 차이 : 매치된 전체 문제와 1번째
    print(url)                      # <a href="(.*)"> 의 뜻은 a href 뒤에 오는 (.*) 아무 문자열이나 가져와서 group으로 만들어라.
```

책 제목 추출
- sub : 대체하는 것 <> --> ''로 대체

```python
import re
from urllib.request import urlopen
from html import unescape

f = urlopen('http://www.hanbit.co.kr/store/books/full_book_list.html')
edcoding = f.info().get_content_charset(failobj='utf-8')
text = f.read().decode(encoding)

for partial_html in re.findall('<td class="left"><a.*</td>',text): # text안에서 '<td> ~ '를 찾으라는 뜻 
#     print(partial_html)
    url = re.search('<a href="(.*)">', partial_html).group(1)
#     print(url)                                
    url = "http://www.hanbit.co.kr" + url
    title = re.sub('<.*?>', '',partial_html)  # pa 에서' <어떤 글자가 와도 좋고, 있든 없든 상관 없다.>' 있는 것을 ''로 바꿔라
    title = unescape(title)          #  이상한 글자들 삭제
#     print(title)                      
    print('url : ', url)
    print('title : ', title)
    print('--')
```

네이버 상품명 추출  답지 - 정규표현식으로 스크레이핑

```python
f = urlopen('https://search.shopping.naver.com/search/category.nhn?cat_id=50000148')
edcoding = f.info().get_content_charset(failobj='utf-8')
text = f.read().decode(encoding)

for partial_html in re.findall('<div class="tit">.*?</div>',text, re.DOTALL): # re.DOTALL : \n이 와도 상관없다는 뜻.
    pro = re.search('title="(.*?)"', partial_html).group(1)  # .*? 사용 추천 검색 용이
    pro = unescape(pro)    
    print('제품명 : ',pro)
    print('--')                           
```
